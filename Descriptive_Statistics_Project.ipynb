```python
# defined function-mean
def mean(list):
    sum=0
    length=len(list)
    for i in range(len(list)):
         sum=sum+list[i]
    mean=sum/length
    return mean
```

```python
#define function median
#defined bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1] :
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                
# defined function median
def median(list):
    bubble_sort(list)
    l = len(list)
    
    mid = (l-1)//2
    
    if(l%2==0):
        return (list[mid] + list[mid+1])/2
    else:
        return list[mid]
  
```

```python
#define variance--using loop 
def var(list):
    sum1=0
    sum2=0
    length=len(list)
    for i in range(len(list)):
        sum1=sum1+list[i]*list[i]
        sum2=sum2+list[i]
    v=(sum1/length)-(sum2/length)**2
    return v
         
    
```

```python
# alternative method of finf=ding variance without using loop and can be applied to numpy array
def variance(X):
    var=mean(X**2)-mean(X)**2
    return var
        
```

```python
# defined function for standard deviation
def s_dev(X):
    s=math.sqrt(mean(X**2)-mean(X)**2)
    return s
```

```python
#for geometric mean
def g_mean(list):
    p=1
    length=len(list)
    for i in range(len(list)):
         p=p*list[i]
    m=(p)**(1/length)
    return m

        
```

```python
#for harmonic mean
def h_mean(list):
    sum=0
    length=len(list)
    for i in range(len(list)):
         sum=sum+(1/list[i])
    hm=length/sum
    return hm
            
    
```

```python
#user defined for covariance
def cova(a,b):
    c=mean(a*b)-mean(a)*mean(b)
    return c
def cov_matrix(a,b):
    L=[cova(a,a),cova(a,b),cova(a,b),cova(b,b)]
    array1=np.array(L)
    array2=array1.reshape(2,2)
    return array2
```
